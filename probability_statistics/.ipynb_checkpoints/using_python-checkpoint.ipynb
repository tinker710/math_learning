## 实验1. 抛硬币实验的模拟

```python
## 
from collections import Counter
import matplotlib.pyplot as plt
import random
times = 10000

count = [] # storage all events
f = []  # storage frequency
indices = [] # list for storaging index
for i in range(1,times+1):
    y = random.randint(0,1)
    count.append(y)
    heads = 0
    for j in range(i):
        if count[j] == 1:
            heads += 1
    f.append(heads/i) # calculate frequency
    indices.append(i)
plt.plot(indices,f)
plt.show()

a = Counter(count)
f1 = a[1]
print(f1/times)
```
